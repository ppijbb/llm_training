```python
from datasets import load_dataset, get_dataset_config_names, concatenate_datasets

data = load_dataset("Gunulhona/open_m_3", "default", split="train", streaming=False)
data = data.shuffle()
data = data.select(range(10))
print(data)
```

```python
from PIL import Image


for idx, images in enumerate(data['images']):
    if len(images) > 0:
        print(idx, images)
        for _image in images:
           _image.show()
```

```python
test_data = data[1]
test_image = test_data["images"][0]
test_data
```

```python
processed_message = []
for idx, (turn, content) in enumerate(zip(test_data["messages"]['role'], test_data["messages"]["content"])):

    processed_content = []
    for type_, content in zip(content["type"], content["text"]):
            processed_content.append({
                "type": type_,
                type_: content if type_ == "text" else test_data["images"].pop(0)
            })

    processed_message.append(
        {
            "role": turn, 
            "content": processed_content
        }
    )
print(processed_message)
```

```python
from transformers import AutoProcessor

with open("/home/conan/workspace/llm_training/sft/config/chat_template.txt", "r") as f:
    chat_template = f.read()

processor = AutoProcessor.from_pretrained("google/gemma-3-4b-it", chat_template=chat_template)

result = processor.apply_chat_template(
    processed_message,
    tokenize=False,
    add_generation_prompt=False
)
print(result)
processor(
    text=result,
    images=test_image,
    image_processor_kwargs={
        "do_resize": True,
        "size": {"height": 896, "width": 896},
        "do_normalize": True,
        "image_mean": [0.5, 0.5, 0.5],
        "image_std": [0.5, 0.5, 0.5]
    })
```
